```python
from urllib.request import urlopen
# https://free-eunb.tistory.com/15

html = urlopen('https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%ED%84%B0%EB%84%B7_%EC%8B%A0%EC%A1%B0%EC%96%B4_%EB%AA%A9%EB%A1%9D')
```

```python
from bs4 import BeautifulSoup
```

```python
bsobject = BeautifulSoup(html, 'html.parser')
```

```python
li_code=bsobject.select('div.mw-parser-output>ul>li')
```

```python
import re
```

```python
pattern = r'\[[0-9]*\]'
```

```python
words = []
means = []

for i in range(len(li_code)-12): # 필요없는 각주 참고 개수 12개
    text = re.sub(pattern=pattern, repl='', string= str(li_code[i].get_text()))
    if ':' in text:
        text = text.replace(u'\xa0', u'') # 인코딩 과정에서의 오류
        index = text.find(':')
        w,m = text[:index].strip(), text[index+1:].strip()
        words.append(w)
        means.append(m)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'word' : words, 'meaning' : means})
df.to_csv('wiki_newwords.csv',index=False,encoding="utf-8-sig")
```
